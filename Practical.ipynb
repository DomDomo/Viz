# Visualizing the Quality of Life

```python
import pandas as pd
import altair as alt
```

```python
source = pd.read_excel('OECD_Data.xlsx')

source.head(10)
```

```python
source.dtypes
```

```python
indicators = list(source.columns)[1:]

indicator_dropdown = alt.binding_select(options=indicators, name="Indicator")

indicator_param = alt.param(value=indicators[0], bind=indicator_dropdown)
```

```python
color_scale = alt.condition(
    alt.datum.Country == 'OECD - Total',  # If the country is 'OECD - Total'
    alt.value('#f58519'), 
    alt.Color('x:Q', scale=alt.Scale(scheme='blues'), legend=None)
)

bar_chart = alt.Chart(source).mark_bar().encode(
    x=alt.X('x:Q', title=''),
    y=alt.Y('Country:N', sort='-x'),
    color=color_scale
).properties(
    width=300,
    height=500
).transform_calculate(
    x=f'datum[{indicator_param.name}]' # Variable X axis
).transform_filter(
    'isValid(datum.x)' # Removes NaN values
).add_params(
    indicator_param # Add dropdown parameter
)

bar_chart
```

```python
color_scale = alt.condition(
    alt.datum.Country == 'OECD - Total',  # If the country is 'OECD - Total'
    alt.value('#f58519'), 
    alt.Color('y:Q', scale=alt.Scale(scheme='blues'), legend=None)
)

alt.Chart(source).mark_bar().encode(
    x=alt.X('Country:N', sort='-y', axis=alt.Axis(labelAngle=-45)),
    y=alt.Y('y:Q', title=''),
    color=color_scale,
    tooltip=[alt.Tooltip('Country:N', title='Country'), alt.Tooltip('y:Q', title='Value')]
).properties(
    width=700,
    height=300
).transform_calculate(
    y=f'datum[{indicator_param.name}]' # Variable Y axis
).transform_filter(
    'isValid(datum.y)' # Removes NaN values
).add_params(
    indicator_param # Add dropdown parameter
)
```
